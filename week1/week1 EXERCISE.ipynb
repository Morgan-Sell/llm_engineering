```python
# imports
import os
import json
from dotenv import load_dotenv
from IPython.display import Markdown, display, update_display
from scraper import fetch_website_links, fetch_website_contents
from openai import OpenAI
```

```python
# constants

MODEL_GPT = 'gpt-4o-mini'
MODEL_LLAMA = 'llama3.2'
openai = OpenAI()
```

```python
# set up environment
load_dotenv(override=True)
api_key = os.getenv("OPENAI_API_KEY")

if api_key and api_key.startswith('sk-proj-') and len(api_key)>10:
    print("API key looks good so far")
else:
    print("There might be a problem with your API key? Please visit the troubleshooting notebook!")
```

```python
# here is the question; type over this to ask something new
question = """
Explain the importance of friction in both the physical and social world.
"""

tutor_system_prompt = """
You're Bill Nye The Science Guy. You make all topics interesting and fun.
People adore you for how you incorporate creativity and comedy in your teachings 
about scientific topics.
"""
```

```python
def stream_tutor(model):
    stream = openai.chat.completions.create(
    model=model,
    messages=[
        {"role": "system", "content": tutor_system_prompt},
        {"role": "user", "content": question}
    ],
    stream=True
    )
    response = ""
    display_handle = display(Markdown(""), display_id=True)
    for chunk in stream:
        response += chunk.choices[0].delta.content or ""
        update_display(Markdown(response), display_id=display_handle.display_id)
```

```python
# Get gpt-4o-mini to answer, with streaming
openai = OpenAI()

stream_tutor(MODEL_GPT)
```

```python
# Get Llama 3.2 to answer
openai = OpenAI(
    base_url="http://localhost:11434/v1",
    api_key="llama3.2"
)

stream_tutor(MODEL_LLAMA)
```
